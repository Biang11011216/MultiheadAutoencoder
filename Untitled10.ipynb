```python
# 在所有 import 之前加上這一行
import os
os.environ["KMP_DUPLICATE_LIB_OK"] = "TRUE"
import cv2
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import numpy as np
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm
import warnings
warnings.filterwarnings('ignore')
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"訓練將使用設備: {device}")
```

```python
# 定義 PyTorch Dataset
class SonarNormalDataset(Dataset):
    def __init__(self, folder_path):
        self.folder_path = folder_path
        # 抓取資料夾內所有的圖檔
        self.image_files = [f for f in os.listdir(folder_path) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
        print(f"找到 {len(self.image_files)} 張正常聲納影像準備訓練。")

    def __len__(self):
        return len(self.image_files)

    def __getitem__(self, idx):
        img_path = os.path.join(self.folder_path, self.image_files[idx])
        # 1. 讀取為灰階
        img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
        
        # 如果讀取失敗 (壞檔)，回傳全黑陣列防呆
        if img is None:
            img = np.zeros((128, 128), dtype=np.float32)
            
        # 2. 縮放與正規化
        img_resized = cv2.resize(img, (128, 128))
        img_normalized = img_resized.astype(np.float32) / 255.0
        
        # 3. 轉為 Tensor [C, H, W] -> [1, 128, 128]
        img_tensor = torch.tensor(img_normalized).unsqueeze(0)
        return img_tensor
```

```python
# 設定資料夾路徑 (請改成你實際的路徑)
TRAIN_FOLDER = "./mix"
BATCH_SIZE = 16

# 建立 Dataset 與 DataLoader
train_dataset = SonarNormalDataset(TRAIN_FOLDER)
train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True, drop_last=True)
```

```python
# Cell 2: 定義 CAE 模型架構
class LightweightSonarAutoencoder(nn.Module):
    def __init__(self):
        super(LightweightSonarAutoencoder, self).__init__()
        # Encoder (壓縮特徵)
        self.encoder = nn.Sequential(
            nn.Conv2d(1, 16, kernel_size=3, stride=2, padding=1),
            nn.ReLU(),
            nn.Conv2d(16, 32, kernel_size=3, stride=2, padding=1),
            nn.ReLU(),
            nn.Conv2d(32, 64, kernel_size=3, stride=2, padding=1),
            nn.ReLU(),
            nn.Conv2d(64, 128, kernel_size=3, stride=2, padding=1) 
        )
        # Decoder (重建影像)
        self.decoder = nn.Sequential(
            nn.ConvTranspose2d(128, 64, kernel_size=3, stride=2, padding=1, output_padding=1),
            nn.ReLU(),
            nn.ConvTranspose2d(64, 32, kernel_size=3, stride=2, padding=1, output_padding=1),
            nn.ReLU(),
            nn.ConvTranspose2d(32, 16, kernel_size=3, stride=2, padding=1, output_padding=1),
            nn.ReLU(),
            nn.ConvTranspose2d(16, 1, kernel_size=3, stride=2, padding=1, output_padding=1),
            nn.Sigmoid() 
        )

    def forward(self, x):
        return self.decoder(self.encoder(x))

model = LightweightSonarAutoencoder().to(device)
```

```python
# Cell 3: 開始訓練模型
EPOCHS = 30           # 訓練輪數 (通常簡單的重建任務 30~50 輪就收斂了)
LEARNING_RATE = 1e-3  # 學習率

# 定義 Loss 與優化器
criterion = nn.MSELoss() 
optimizer = optim.AdamW(model.parameters(), lr=LEARNING_RATE)

history_loss = []
best_loss = float('inf')

print("🚀 開始訓練 Autoencoder...")

for epoch in range(EPOCHS):
    model.train()
    running_loss = 0.0
    
    # 建立進度條
    progress_bar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{EPOCHS}", leave=False)
    
    for batch_images in progress_bar:
        batch_images = batch_images.to(device)
        
        # 1. 前向傳播 (Forward)
        reconstructed = model(batch_images)
        
        # 2. 計算 Loss (Input vs Output)
        loss = criterion(reconstructed, batch_images)
        
        # 3. 反向傳播與優化 (Backward & Optimize)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        running_loss += loss.item()
        progress_bar.set_postfix({"loss": f"{loss.item():.5f}"})
        
    # 計算這個 Epoch 的平均 Loss
    avg_loss = running_loss / len(train_loader)
    history_loss.append(avg_loss)
    print(f"Epoch [{epoch+1}/{EPOCHS}] | Average MSE Loss: {avg_loss:.5f}")
    
    # 儲存表現最好的模型權重
    if avg_loss < best_loss:
        best_loss = avg_loss
        torch.save(model.state_dict(), 'best_autoencoder.pth')
        print(f"  🌟 找到更低 Loss，已儲存模型至 'best_autoencoder.pth'")

print("✅ 訓練完成！")
```

```python
# Cell 3: 定義強化版單幀影像推論函數 (修正暗圖偏誤)
criterion = nn.MSELoss()

def process_frame(frame, model, threshold_rel=0.065, signal_p98_threshold=0.45, texture_threshold=500):
    """
    三重驗證機制：
    1. P99 百分位數：排除零星噪點 (空發) 與 全黑畫面
    2. Texture Score：排除平滑故障畫面 (藍屏/系統訊息)
    3. Relative MSE：透過 AE 識別結構性異常
    """
    # 1. 影像前處理
    gray_img = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    # --- A. 計算紋理複雜度 (Laplacian) ---
    texture_score = cv2.Laplacian(gray_img, cv2.CV_64F).var()
    
    img_resized = cv2.resize(gray_img, (128, 128))
    img_normalized = img_resized.astype(np.float32) / 255.0
    
    # --- B. 計算百分位亮度 (P98) ---
    # 解決側掃 Nadir 太暗與空發亮點問題。若 P98 太低，代表畫面沒有成片的有效信號
    p98_val = np.percentile(img_normalized, 98) 
    avg_brightness = img_normalized.mean()
    
    # 2. 模型推論 (AE 重建)
    input_tensor = torch.tensor(img_normalized).unsqueeze(0).unsqueeze(0).to(device)
    with torch.no_grad():
        reconstructed = model(input_tensor)
        mse_loss = nn.MSELoss()(reconstructed, input_tensor).item()
    
    # --- C. 計算相對重建誤差 ---
    relative_score = mse_loss / (avg_brightness + 1e-6)
    
    # --- 核心判定邏輯升級 ---
    # 1. 判斷是否完全沒訊號 (包含空發的零星亮點)
    is_no_signal = p98_val < signal_p98_threshold
    
    # 2. 判斷是否為平滑故障圖 (如藍屏)
    is_too_flat = texture_score < texture_threshold
    
    # 3. 判斷是否為 AI 認定之結構異常
    is_high_error = relative_score > threshold_rel
    
    # 綜合判定
    is_anomaly = is_no_signal or is_too_flat or is_high_error
    
    # 判定原因分類顯示
    if is_no_signal: reason = "NO SIGNAL / AIR"
    elif is_too_flat: reason = "SYSTEM FLAT ERROR"
    elif is_high_error: reason = "AE RECON ERROR"
    else: reason = "VALID"

    # 4. 繪製結果
    output_frame = frame.copy()
    text_color = (0, 0, 255) if is_anomaly else (0, 255, 0)
    
    # 在畫面上顯示多重指標
    cv2.putText(output_frame, f"Status: {reason}", (30, 50), 
                cv2.FONT_HERSHEY_SIMPLEX, 1.0, text_color, 2, cv2.LINE_AA)
    
    cv2.putText(output_frame, f"P98: {p98_val:.2f} | Texture: {texture_score:.1f}", (30, 90), 
                cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2, cv2.LINE_AA)

    cv2.putText(output_frame, f"Rel MSE: {relative_score:.4f}", (30, 130), 
                cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2, cv2.LINE_AA)
    
    if is_anomaly:
        h, w = output_frame.shape[:2]
        cv2.rectangle(output_frame, (0, 0), (w-1, h-1), (0, 0, 255), 10)
        
    return output_frame, relative_score
```

```python
# Cell 4: 處理整段影片
INPUT_VIDEO_PATH = "valf/gt/2026_03_19_15_55_15_3_video.mp4"  # 替換成你的測試影片路徑
OUTPUT_VIDEO_PATH = "anomaly_result.mp4" # 輸出的結果影片名稱
ANOMALY_THRESHOLD = 0.07                # 💡 請依據你實際資料的數值微調此閾值

cap = cv2.VideoCapture(INPUT_VIDEO_PATH)

if not cap.isOpened():
    print("❌ 無法開啟影片，請檢查路徑是否正確！")
else:
    # 獲取影片屬性
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = cap.get(cv2.CAP_PROP_FPS)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    
    # 設定影片寫入器 (使用 mp4v 編碼)
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(OUTPUT_VIDEO_PATH, fourcc, fps, (width, height))
    
    print(f"🎬 開始處理影片: {INPUT_VIDEO_PATH}")
    print(f"解析度: {width}x{height}, 總幀數: {total_frames}, FPS: {fps}")
    
    # 紀錄所有的 MSE 以供後續分析
    history_mse = []
    
    # 使用 tqdm 顯示處理進度條
    for _ in tqdm(range(total_frames), desc="影片處理進度"):
        ret, frame = cap.read()
        if not ret:
            break
            
        # 進行異常偵測並畫圖
        processed_frame, frame_mse = process_frame(frame, model, threshold_rel=ANOMALY_THRESHOLD)
        history_mse.append(frame_mse)
        
        # 寫入新影片
        out.write(processed_frame)
        
    # 釋放資源
    cap.release()
    out.release()
    print(f"✅ 處理完成！結果已儲存至: {OUTPUT_VIDEO_PATH}")
```

```python
import cv2
import os
from tqdm import tqdm

def split_video_to_10s(input_path, output_folder):
    # 1. 建立輸出資料夾
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
        print(f"建立資料夾: {output_folder}")

    # 2. 開啟影片
    cap = cv2.VideoCapture(input_path)
    if not cap.isOpened():
        print("❌ 無法開啟影片檔案")
        return

    # 3. 獲取影片屬性
    fps = cap.get(cv2.CAP_PROP_FPS)
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    
    # 計算 10 秒有多少幀
    frames_per_clip = int(fps * 10)
    
    print(f"🎬 影片屬性: {width}x{height} | FPS: {fps:.2f} | 總長度: {total_frames/fps:.2f} 秒")
    print(f"📦 每 {frames_per_clip} 幀切分一次 (10秒)")

    clip_count = 29
    frame_count = 0
    out = None
    
    # 建立進度條
    pbar = tqdm(total=total_frames, desc="影片切分進度")

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        # 如果是一段新 Clip 的開始
        if frame_count % frames_per_clip == 0:
            if out is not None:
                out.release() # 關閉前一個 Clip
            
            output_name = os.path.join(output_folder, f"clip_{clip_count:03d}.mp4")
            fourcc = cv2.VideoWriter_fourcc(*'mp4v') # 使用 mp4 格式
            out = cv2.VideoWriter(output_name, fourcc, fps, (width, height))
            clip_count += 1

        # 寫入影格
        out.write(frame)
        
        frame_count += 1
        pbar.update(1)

    # 釋放資源
    if out is not None:
        out.release()
    cap.release()
    pbar.close()
    print(f"✅ 切分完成！共生成 {clip_count} 個片段。")

# --- 使用範例 ---
split_video_to_10s("C:/Users/user/Desktop/yoloversion/test/2026_05_06_11_34_33_2_video.mp4", "C:/Users/user/Desktop/yoloversion/test")
```

```python
import cv2
import os
from tqdm.notebook import tqdm

def extract_frames(video_folder, output_folder, frame_stride=5):
    """
    將所有影片的影格提取後，統一放在同一個資料夾中
    video_folder: 影片來源資料夾
    output_folder: 影像統一儲存的目標資料夾
    frame_stride: 每隔幾幀抓一張
    """
    # 建立唯一的輸出資料夾
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
        print(f"建立統一儲存資料夾: {output_folder}")
        
    video_files = [f for f in os.listdir(video_folder) if f.lower().endswith(('.mp4', '.avi', '.mov'))]
    print(f"找到 {len(video_files)} 部影片，準備開始提取並匯總影像...")

    total_extracted = 0

    for video_name in tqdm(video_files, desc="處理影片中"):
        video_path = os.path.join(video_folder, video_name)
        video_id = os.path.splitext(video_name)[0] # 取得影片檔名作為前綴
        
        cap = cv2.VideoCapture(video_path)
        frame_count = 0
        saved_in_this_video = 0
        
        while True:
            ret, frame = cap.read()
            if not ret:
                break
            
            # 依照 Stride 進行抽樣
            if frame_count % frame_stride == 0:
                # 檔名格式: 影片名_影格號.jpg (確保唯一性)
                img_name = f"{video_id}_f{frame_count:06d}.jpg"
                save_full_path = os.path.join(output_folder, img_name)
                
                # 直接儲存到統一資料夾
                cv2.imwrite(save_full_path, frame, [cv2.IMWRITE_JPEG_QUALITY, 95])
                saved_in_this_video += 1
            
            frame_count += 1
            
        cap.release()
        total_extracted += saved_in_this_video

    print(f"✅ 提取完成！")
    print(f"📁 目錄路徑: {output_folder}")
    print(f"🖼️ 總影像張數: {total_extracted}")


# --- 使用範例 ---
extract_frames("clip/", "./train_images/", frame_stride=15)
```

```python
import os
import cv2
import torch
import shutil
import time
import numpy as np
from tqdm.notebook import tqdm

# --- 1. 設定路徑與門檻 ---
SOURCE_FOLDER = "./2026_03_20/"  # 原始影片存放資料夾
NORMAL_FOLDER = "./2026_03_20/"  # 篩選出的正常影片存放處
ANOMALY_FOLDER = "./f/"# 篩選出的異常影片存放處

# 門檻同步 (依據你 process_frame 的調校值)
SIGNAL_P98_THRESHOLD = 0.45  # P98 百分位亮度：低於此值代表沒訊號 (或只有零星噪點)
TEXTURE_MIN = 500.0          # 紋理變異數：低於此值代表畫面太平滑 (藍屏/故障)
REL_SCORE_MAX = 0.065        # 相對重建誤差：高於此值為 AI 判定之結構異常
TRIGGER_PERCENT = 0.10       # 影片中超過 10% 的幀異常，則判定整部影片異常
ANOMALY_THRESHOLD = 0.07
# 建立資料夾
os.makedirs(NORMAL_FOLDER, exist_ok=True)
os.makedirs(ANOMALY_FOLDER, exist_ok=True)

def classify_single_video(video_path, model):
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        return None, 0
    
    anomaly_frame_count = 0
    processed_frame_count = 0
    frame_stride = 5 # 每 5 幀抽樣一次，提升分類速度
    
    try:
        while True:
            ret, frame = cap.read()
            if not ret:
                break
            
            if processed_frame_count % frame_stride == 0:
                # 1. 影像前處理
                gray_img = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
                img_resized = cv2.resize(gray_img, (128, 128))
                img_normalized = img_resized.astype(np.float32) / 255.0
                
                # --- A. 百分位亮度檢查 (解決側掃 Nadir 與 空發亮點) ---
                # P99 代表畫面中前 1% 的亮度水平，能有效過濾掉孤立的脈衝噪點
                p99_val = np.percentile(img_normalized, 98) 
                is_no_signal = p99_val < SIGNAL_P98_THRESHOLD
                
                # --- B. 紋理複雜度檢查 (Laplacian Variance) ---
                texture_score = cv2.Laplacian(gray_img, cv2.CV_64F).var()
                is_too_flat = texture_score < TEXTURE_MIN
                
                # --- C. 相對重建誤差檢查 (R-CAE) ---
                is_high_error = False
                avg_brightness = img_normalized.mean()
                
                # 策略優化：只有在「有訊號」且「不是藍屏」的情況下，才跑耗能的 AI 推論
                if not is_no_signal and not is_too_flat:
                    input_tensor = torch.tensor(img_normalized).unsqueeze(0).unsqueeze(0).to(device)
                    with torch.no_grad():
                        recon = model(input_tensor)
                        mse = torch.nn.functional.mse_loss(recon, input_tensor).item()
                        rel_score = mse / (avg_brightness + 1e-6)
                    
                    # if rel_score > REL_SCORE_MAX:
                    #     is_high_error = True
                
                
                # 三重驗證：任何一項不通過即視為異常影格
                if is_no_signal or is_too_flat or is_high_error:
                    anomaly_frame_count += 1
                            
            processed_frame_count += 1
    finally:
        cap.release() # 🛡️ 釋放影片，避免 Windows WinError 32
    
    # 計算該影片總共抽樣了多少幀
    actual_sampled_count = (processed_frame_count + frame_stride - 1) // frame_stride
    anomaly_ratio = anomaly_frame_count / (actual_sampled_count + 1e-6)
    
    return anomaly_ratio > TRIGGER_PERCENT, anomaly_ratio

# --- 2. 執行批次掃描與檔案搬移 ---
# 解決 OpenMP 衝突問題
os.environ["KMP_DUPLICATE_LIB_OK"] = "TRUE"

video_files = [f for f in os.listdir(SOURCE_FOLDER) if f.lower().endswith(('.mp4', '.avi', '.mov'))]
print(f"🚀 開始批次分類，共計 {len(video_files)} 部影片...")

for file_name in tqdm(video_files, desc="分類進度"):
    file_path = os.path.join(SOURCE_FOLDER, file_name)
    
    # 進行判定
    is_anomaly, ratio = classify_single_video(file_path, model)
    
    if is_anomaly is None:
        continue
    
    # 決定目標路徑
    target_dir = ANOMALY_FOLDER if is_anomaly else NORMAL_FOLDER
    target_path = os.path.join(target_dir, file_name)
    
    # 🛡️ 檔案搬移前的緩衝與保護
    time.sleep(0.1) 
    try:
        shutil.move(file_path, target_path)
    except PermissionError:
        time.sleep(1.0) # 被鎖定時多等一秒再重試
        try:
            shutil.move(file_path, target_path)
        except Exception as e:
            print(f"❌ 無法搬移 {file_name}，檔案可能正被其他程式開啟: {e}")

print(f"✅ 分類完成！")
print(f"正常影片存放在: {NORMAL_FOLDER}")
print(f"異常影片存放在: {ANOMALY_FOLDER}")
```
